#### **Como construir tensores com TensorFlow?**

Vamos começar a escrever algum código. Se você não tiver o TensorFlow instalado, poderá usar um bloco de anotações do [Google Colab](https://colab.research.google.com/) para acompanhar.

Vamos começar importando o TensorFlow e imprimindo a versão

```python
import tensorflow as tf

print(f'[VERSION] TensorFlow: {tf.__version__}')
```

Vamos primeiro criar um escalar usando tf.constant. Usamos tf.constant para criar um novo valor constante. Também podemos usar tf.variable para criar um valor de variável. Em seguida, imprimiremos o valor e também verificaremos a dimensão do escalar usando a propriedade ndim. Sua dimensão será zero porque é um valor único.

```python
scalar = tf.constant(7)

print(scalar, end='\n\n')
print(scalar.ndim)
```

Agora vamos criar um vetor e imprimir suas dimensões. Você pode ver que a dimensão é 1.

```python
vector = tf.constant([10,10])

print(vector, end='\n\n')
print(vector.ndim)
```

Agora vamos tentar criar uma matriz e imprimir suas dimensões.

```python
matrix = tf.constant([
    [10,11],
    [12,13]
])

print(matrix, end='\n\n')
print(matrix.ndim)
```

Você verá que a dimensão agora é 2. Você também pode ver que a forma da matriz é 2 por 2.

Formas e dimensões são úteis ao trabalhar com o TensorFlow porque frequentemente as alteramos ao usar esses dados para treinar redes neurais.

Vimos que esses tensores têm um tipo de dados padrão de int32. E se quisermos criar um conjunto de dados com um tipo de dados personalizado?

tf.constant nos fornece o argumento dtype. Vamos criar a mesma matriz novamente com float16 como tipo de dado.

```python
tensor = tf.constant([
    [
        [1,2,3]
    ],
    [
        [4,5,6]
    ],
    [
        [7,8,9]
    ]
], dtype='float32')

print(tensor, end='\n\n')
print(tensor.ndim)
```

Agora vamos criar um tensor. Vamos inserir um array tridimensional em tf.constant. Também imprimiremos suas dimensões.

```python
tensor = tf.constant([
    [
        [1,2,3]
    ],
    [
        [4,5,6]
    ],
    [
        [7,8,9]
    ]
])

print(tensor, end='\n\n')
print(tensor.ndim)
```

Agora temos um tensor de dimensão 3 e forma 3 por 1 por 3. Este é o tensor mais simples que você pode criar. Em cenários do mundo real, estaremos lidando com tensores de dimensões maiores e formas maiores.

Agora vamos ver como criar um tensor variável. Não usaremos tensores variáveis ​​com muita frequência em comparação com tensores constantes, mas é bom saber que temos uma opção.

Usaremos tf.Variable para criar um tensor variável. A diferença entre o tensor constante e o tensor variável é que você pode alterar os dados em um tensor variável, mas não pode alterar os valores em um tensor constante. Vamos criar um tensor variável e imprimir as dimensões.

```python
var_tensor = tf.Variable([
    [
        [1,2,3]
    ],
    [
        [4,5,6]
    ],
    [
        [7,8,9]
    ]
])

print(var_tensor)
```

#### **Como gerar e carregar tensores**

Vejamos como gerar tensores. Na maioria dos casos, você não criará tensores do zero. Você carregará um conjunto de dados, converterá outros conjuntos de dados como matrizes `NumPy` em tensores ou gerará tensores. Primeiro, vamos ver como gerar tensores.

Vamos criar um tensor com valores aleatórios. Existem duas maneiras comuns de fazer isso: gerar uma distribuição normal de dados ou uma distribuição uniforme de dados.

image.png

<center>Distribuição normal</>

A _distribuição normal_ é uma curva em forma de sino que representa a distribuição dos dados. A maioria dos dados estará próxima da média e poucos dados estarão distantes da média. Isso significa que a probabilidade de obter um valor próximo à média é maior.

image.png

<center>Distribuição uniforme</>

A distribuição uniforme é uma linha reta que representa a distribuição dos dados. Todos os valores em uma distribuição uniforme terão uma probabilidade igual de ocorrer dentro de um determinado intervalo.

Antes de gerarmos valores aleatórios, você deve entender o que é uma semente. Se usarmos um valor de semente, podemos gerar novamente o mesmo conjunto de dados várias vezes. Isso será útil quando quisermos testar nosso modelo de aprendizado de máquina com os mesmos dados depois de ajustarmos seu desempenho.

Vamos criar dois arrays de tensores aleatórios. Primeiro, definiremos uma semente e geraremos os valores aleatórios usando essa semente.

```python
seed = tf.random.Generator.from_seed(42)
```

Agora vamos criar uma distribuição normal e uniforme com a forma de 3 por 2.

```python
normal_tensor = seed.normal(shape=(3,2))

print(normal_tensor)
```

```python
uniform_tensor = seed.uniform(shape=(3,2))

print(uniform_tensor)
```

Temos dois tensores criados, um com distribuição normal de números aleatórios e outro com distribuição uniforme de números aleatórios.

Em seguida, criaremos um tensor com zeros e uns. No TensorFlow, os tensores preenchidos com zeros ou uns costumam ser usados ​​como ponto de partida para a criação de outros tensores. Eles também podem ser espaços reservados para entradas em um gráfico computacional.

Para criar um tensor de zeros, use a função tf.zeros com uma forma como argumento de entrada. Para criar um tensor com uns, usamos tf.ones com a forma como argumento de entrada.

```python
zeros = tf.zeros(shape=(3,2))

print(zeros)
```

```python
ones = tf.ones(shape=(3,2))

print(ones)
```

Agora, vamos ver como converter matrizes `NumPy` em tensores. Se você não sabe o que é [NumPy](https://numpy.org/), é uma biblioteca Python para computação numérica. Ele nos ajuda a lidar com grandes conjuntos de dados e realizar uma variedade de cálculos neles.

Vamos importar o NumPy e criar um array NumPy usando a função de arranjo do NumPy.

```python
import numpy as np

numpy_arr = np.arange(1,25, dtype=np.int32)

print(numpy_arr)
```

Agora, podemos criar um tensor usando a função tf.constant com o array NumPy como entrada. O TensorFlow possui suporte integrado para lidar com matrizes NumPy, portanto, é apenas uma questão de importar uma matriz NumPy e definir uma forma.

```python
numpy_tensor = tf.constant(numpy_arr, shape=[2,4,3])

print(numpy_tensor)
```

Você pode ver tanto o array NumPy quanto nosso tensor. A matriz NumPy original era 1x12, mas nosso tensor é 2x4x3. Isso é chamado de remodelar um tensor, o que geralmente fazemos durante o treinamento de redes neurais profundas.

#### **Operações básicas usando o Tensorflow**

Aprendemos como os tensores são criados no TensorFlow. Agora vamos ver algumas operações básicas usando tensores.

Começaremos obtendo algumas informações sobre nossos tensores. Vamos criar um tensor 4D com valores 0 com a forma 2x3x4x5.

```python
rank4_tensor = tf.zeros(
    [2,3,4,5]
)

print(rank4_tensor)
```

Criamos nosso tensor de classificação 4. Agora vamos obter algumas informações sobre o tamanho, forma (número de valores) e a dimensão do tensor.

Usaremos a função tf.size para obter o tamanho. As propriedades de forma e ndim nos darão a forma e as dimensões do tensor.

```python
print(f'Size     : {tf.size(rank4_tensor)}')
print(f'Shape    : {rank4_tensor.shape}')
print(f'Dimension: {rank4_tensor.ndim}')
```

Vejamos alguns cálculos simples usando o tensor. Vou criar um novo tensor básico.

```python
basic_tensor = tf.constant([
    [10,11],
    [12,13]
])

print(basic_tensor)
```

Vamos tentar algumas operações simples. Podemos adicionar, subtrair, multiplicar e dividir todos os valores em um tensor usando os operadores básicos.

```python
print(basic_tensor + 10)
```

```python
print(basic_tensor - 10)
```

```python
print(basic_tensor * 10)
```

```python
print(basic_tensor / 10)
```

Agora vamos tentar a multiplicação de matrizes. Vou criar dois tensores simples tensor_01 e tensor_02.

```python
tensor_01 = tf.constant([
    [2,2], [4,4]
])

tensor_02 = tf.constant([
    [2,3], [4,5]
])
```

Lembre-se de que, na multiplicação de matrizes, as dimensões internas devem corresponder. Por exemplo, uma multiplicação (3, 5) * (3, 5) não funcionará, mas (3, 5) * (5, 3) funcionará.

A forma final da matriz resultante será sua dimensão externa. então, um tensor 3x5 multiplicado por um tensor 5x3 nos dará um tensor 5x5. Usaremos a função tf.matmul para realizar a multiplicação de matrizes.

```python
print(tensor_01, end='\n\n')
print(tensor_02)
```

```python
print(tf.matmul(tensor_01, tensor_02))
```

A seguir, veremos como remodelar e transpor uma matriz. Como vimos antes, frequentemente usaremos remodelagem para alterar nossa estrutura de matriz durante o treinamento de redes neurais.

Por exemplo, uma matriz de pixel de imagem de 28x28 será convertida em uma matriz unidimensional de 784 pixels para uma rede neural de classificação de imagem.

Para remodelar, usamos a função tf.reshape. Para transpor, usamos a função tf.transpose. Se você não sabe o que é uma transposição, ela converte linhas em colunas e colunas em linhas.

```python
print(tf.reshape(tensor_01, [4,1]), end='\n\n')
print(tf.transpose(tensor_01))
```

Finalmente, vamos ver algumas operações agregadas como min, max, desvio padrão, quadrado e raiz quadrada.

Para encontrar os valores mínimo e máximo, usamos as funções tf.reduce_min e tf.reduce_max. E para encontrar a soma do array, usamos a função tf.reduce_sum.

```python
tensor_03 = tf.constant([
    [1,2,3],
    [4,5,6],
    [7,8,9]
], dtype='float32')

tensor_03
```

```python
print(
    tf.reduce_min(tensor_03),
    tf.reduce_max(tensor_03),
    tf.reduce_sum(tensor_03),
    sep='\n'
)
```

Agora, para o desvio padrão e a variância, usamos a função tf.math.reduce_std e a função tf.math.reduce_variance.

```python
print(
    tf.math.reduce_std(tensor_03),
    tf.math.reduce_variance(tensor_03),
    sep='\n'
)
```

Vamos encontrar o quadrado, a raiz quadrada e o logaritmo de cada valor em um tensor.

```python
print(
    tf.sqrt(tensor_03),
    tf.square(tensor_03),
    tf.math.log(tensor_03),
    sep='\n\n'
)
```
